```python
url = 'https://raw.githubusercontent.com/alura-cursos/pandas-conhecendo-a-biblioteca/main/base-de-dados/aluguel.csv'
dados = pd.read_csv(url, sep=';') #a planilha não estava separada por vírgula por padrão
dados.head()
```

```python
dados['Valor_por_mes'] = dados['Valor'] + dados['Condominio']
dados.head()
```

```python
dados['Valor_por_ano'] = dados['Valor']*12 + dados['IPTU']
dados.head()
```

```python
dados['Descição'] = dados['Tipo'] + ' em ' + dados['Bairro'] + ' com ' + dados['Quartos'].astype(str) + ' quarto(s) ' + ' e ' + dados['Vagas'].astype(str) + ' vaga(s) de garagem.'
dados.head()
```

```python
dados['Possui_suite'] = dados['Suites'].apply(lambda x: "Sim" if x > 0 else "Não")
#método apply aplica determinada funções em alguns dados, que foi dita acima, usando a função lambda que permite escrever em uma únida linha de código, o x representa a cada dado da linha da tabela da coluna suite
dados.head()
```
